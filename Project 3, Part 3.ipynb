### Title_Basics Table

```python
#Check title_basics column names for accuracy in normalization steps
title_basics.head()
```

```python
#Create a column with list of genres
title_basics['genres_split'] = title_basics['genres'].str.split(',')
title_basics
```

```python
#Separate list of genres into new rows
exploded_genres = title_basics.explode('genres_split')
exploded_genres
```

```python
#Sort genres alphabetically
unique_genres = sorted(exploded_genres['genres_split'].unique())
```

```python
#Create title_genres table
title_genres = exploded_genres[['tconst', 'genres_split']].copy()
title_genres.head()
```

```python
#Create genre mapper dictionary
genre_ints = range(len(unique_genres))
genre_map = dict(zip(unique_genres, genre_ints))
genre_map
```

```python
#Make new integer column in title_genres table
title_genres['genre_id'] = title_genres['genres_split'].map(genre_map)
title_genres = title_genres.drop(columns = 'genres_split')
title_genres.head()
```

```python
#Convert genre map dictionary into genres table
genres = pd.DataFrame({'Genre_Name': genre_map.keys(),
                      'Genre_ID': genre_map.values()})
genres.head()
```
